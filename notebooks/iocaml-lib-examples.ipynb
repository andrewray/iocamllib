# TyXML

TyXML allows the generation of _typed_ html fragments - that is the html code is compile time checked making it much harder to write nonsense.

The tyxml.syntax package includes a number of quotation expanders; `<:html<>> <:html5list<>> <:svg<>>` etc to help writing html code.

HTML printers are installed for the types `Html5.M.elt` and `Svg.M.elt`.

```ocaml
#require "tyxml.syntax"
```

```ocaml
module Html5 = Html5.M
```

```ocaml
let _ = <:html5list< <i>hello</i> <b>world</b> >> ;;
```

```ocaml
module Svg = Svg.M
```

```ocaml
let _ = <:svg< <svg> </svg> >>
```

# Caml-on-the-Web

Cow is similar to TyXML, but not quite as strict.  It also provides quotation exanders for html.  It also has a useful css quotation expander and ways of automatically getting html renderers for user types.

```ocaml
#require "cow.syntax"
```

```ocaml
open Cow
```

```ocaml
<:html< <b>hello</b> <i>world</i> >>
```

```ocaml
type x = { a : int; b : float } with html,json;;
```

```ocaml
<:css<
#special {
    .a { color: orange; }
    .b { color: green; }
}
>>
```

```ocaml
html_of_x ~id:"special" { a=3; b=0.3 };;
```

```ocaml
html_of_x ~id:"normal" { a=3; b=0.3 };;
```

```ocaml
json_of_x { a=4; b=0.0 }
```

```ocaml
Cow.Markdown.of_string "
# Hello

I am some markdown

* _With_
* __a__
* ___list___
"
```
